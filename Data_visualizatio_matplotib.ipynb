## Bar Chart
categorical ,

```python
data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
#gender wisw
gender_group=df.groupby('Gender').size()
gender_group.index

```

```python
gender_group.values
```

```python
#y=gender count ,x=male female category
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('bar chart of gender')
plt.bar(gender_group.index,gender_group.values,color=['blue','green'])
```

```python
# bar char on real 
df=pd.read_csv('student_data.csv')
df
```

```python
status=df.groupby('CompletionStatus').size()
plt.xlabel('Completion Status')
plt.ylabel('Count')
plt.title('bar chart of course completion')
plt.bar(status.index,status.values, edgecolor='black')
```

```python
status=df.groupby('Location').size()
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('bar chart of location ')
plt.bar(status.index,status.values, edgecolor='black')
```

## Pie Char

```python
status=df.groupby('CompletionStatus').size()
plt.pie(status,labels=status.index,autopct='%1.1f%%',explode=(0.1,0,0))
plt.show()
```
