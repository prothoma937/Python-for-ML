NdArray

```python
import numpy as np
#1d array

arr1=np.array([1,2,3,4,5])
print(type(arr1))
print(arr1.ndim)

#2d array
arr2=np.array([[1,2,3,4],[5,6,7,8]])
print(arr2.ndim)

#3D ARRAY
arr3=np.array([
    [[1,2,3,4],[5,6,7,8]],
    [[11,12,13,14],[15,16,17,18]],
    [[1,2,3,4],[5,6,7,8]],
])
print(arr3.ndim)
```

Array attribute

```python
#dimension
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
#shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)
#size
print(arr1.size)
print(arr2.size)
print(arr3.size)
#dtypea
print(arr1.dtype)
print(arr2.dtype)
print(arr2.dtype)
```

ndarray creation from range

```python
#np.linspace(start,end(include),koita)

arr=np.linspace(1,10,15)
print(arr)
print(arr.dtype)
print(type(arr))
print(arr.ndim)
print(arr.shape)
print(arr.size)
```

```python
#np.longspace(start,end,koita) power

arr=np.logspace(1,4,5,base=2)
print(arr)
print(arr.dtype)
print(type(arr))
print(arr.ndim)
print(arr.shape)
print(arr.size)
```
