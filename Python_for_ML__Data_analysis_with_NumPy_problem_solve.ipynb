```python
import numpy as np

def calculate_matrix_mean(matrix_data, mode):
    if mode == 'row':
        return list(map(lambda row: float(np.mean(row)), matrix_data))
    elif mode == 'column':
        return list(map(lambda col: float(np.mean(col)), zip(*matrix_data)))

matrix_data = [[1, 2, 3],
               [4, 5, 6],
               [7, 8, 9]]

print("Column Means:", calculate_matrix_mean(matrix_data, mode='column'))
print("Row Means:   ", calculate_matrix_mean(matrix_data, mode='row'))
```

#Matrix-Vector Dot Product

```python
import numpy as np

def matrix_dot_vector(a,b):
    if len(a[0]) != len(b):
        return -1

    return list(map(lambda row: float(np.dot(row, b)), a))
```
