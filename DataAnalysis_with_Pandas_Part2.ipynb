```python
mn=df1['Total Marks'].min()
mx=df1['Total Marks'].max()
df1['Scaled Marks']=df1['Total Marks'].apply(lambda x: (x-mn)/(mx-mn))
df1
```

```python
#custom built function
def grading_system(marks): 
    if marks >=260: 
        return 'A+'
    elif marks>=250:
        return 'A'
    else:
        return 'A-'
df1['Grade']=df1['Total Marks'].apply(grading_system)
df1
```

```python
#axis = 0 মানে হলো রো (Row) বা উপর-নিচ অপারেশন।
#axis = 1 মানে হলো কলাম (Column) বা পাশাপাশি অপারেশন।
def marking_system(df): 
    a=df['Data Structure Marks']*2
    b=df['Algorithm Marks'] * 3  
    c=df['Python Marks'] * 4 
    return a+b+c    


df1['Exceptional Marks']=df1.apply(marking_system, axis=1)
df1
# def marking_system(row): 
#     a = row['Data Structure Marks'] * 2
#     b = row['Algorithm Marks'] * 3  
#     c = row['Python Marks'] * 4 
#     return a + b + c    
# df1['Exceptional Marks'] = df1.apply(marking_system, axis=1)
# df1
```
