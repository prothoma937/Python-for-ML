```python

#positional arguments
def squre_add(a,b):
  print(f"inside value {a} {b}")
  a=a**2
  b=b**2
  return a+b

ans=squre_add(10,30)+10
print(ans)
```

```python
#keyword arguments
def squre_add(a,b,c):
  print(f"inside value {a} {b} {c}")
  a=a**2
  b=b**2
  c=c**2
  return a+b+c

ans=squre_add(a=10,b=20,c=30)+10
print(ans)
```

```python
#variable size args
def squre_add(*args):
#  print(args)
 return args

ans=squre_add(10,20,30)
print(ans)
```

```python
def squre_add(*args):
 return sum(args)

ans=squre_add(10,20,30,40)+10
print(ans)
```

```python
def squre_add(*args):
 for i in args:
  print(i)
 return args

ans=squre_add(10,20,30,40)
print(ans)
```

```python
def squre_add(*args):
 suma=0
 for i in args:
  i=i**2
  suma+=i
 return suma

ans=squre_add(10,20,30,40,50)
print(ans)
```

#generator chunk a separate kore kaj  korbo yield


```python
lst=[x for x in range (500)]

def data_loader(chunk_size,lst):
  for i in range(0,len(lst),chunk_size):
    yield lst[i:i+chunk_size]

x=data_loader(4,lst)
print(next(x))
print(next(x))
print(next(x))
print(next(x))
```

```python
dataset=[x for x in range(500)]
def data_l(chunk_size,dataset):
 for i in range(0,len(dataset),chunk_size):
  yield dataset[i:i+chunk_size]


x=data_l(6,dataset)
print(next(x))
print(next(x))
```
